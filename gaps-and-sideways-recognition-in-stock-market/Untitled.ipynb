```python
atr_multiple = 1.5
n = 20

hist[['avehigh', 'avelow']] = hist[['High', 'Low']].rolling(n).mean()
hist['avemidprice'] = (hist['avehigh'] + hist['avelow']) / 2
```

```python
hist.head()
```

```python
indicator_atr = ta.volatility.AverageTrueRange(hist['High'], hist['Low'], hist['Close'], window=4, fillna=False)

hist['atr14'] = indicator_atr.average_true_range()
hist['UPB'] = hist['avemidprice'] + atr_multiple * hist['atr14']
hist['LOB'] = hist['avemidprice'] - atr_multiple * hist['atr14']
```

```python
hist['rangemaxprice'] = hist[['High']].rolling(n).max()
hist['rangeminprice'] = hist[['Low']].rolling(n).min()
hist['sideways'] = 0
```

```python
def sideways_range(maxp, minp, uppl, lopl):
    if maxp < uppl and maxp > lopl and minp < uppl and minp > lopl:
        return 1
    else:
        return 0
```

```python
hist['sideways'] = hist[['rangemaxprice', 'rangeminprice', 'UPB', 'LOB']].apply(lambda x: sideways_range(x['rangemaxprice'], x['rangeminprice'], x['UPB'], x['LOB'] ), axis=1)
```
